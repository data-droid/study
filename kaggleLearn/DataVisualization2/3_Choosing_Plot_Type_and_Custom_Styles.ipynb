### Trends
>* sns.lineplot : 시간에 따른 트렌드 보기엔 딱!
### Relationship
>* sns.barplot : 여러 그룹간에 수량을 비교하는데 딱!
>* sns.heatmap
>* sns.scatterplot : 두 연속 변수 사이의 관계! 색까지 추가하면 3번째 범주 형 변수와의 관계까지!
>* sns.regplot : scatterplot에 regression line까지 추가!
>* sns.lmplot : multiple regression line에 딱!
>* sns.swarmplot : categorical scatter plots!
### Distribution
>* sns.distplot : 1개의 정수형 변수에 대한 Historgram
>* sns.kdeplot : 1개 또는 2개의 정수형 변수!
>* sns.jointplot : 각 개별 변수에 대한 KDE plot + 2D KDE plot

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
spotify = pd.read_csv('../../../Data/data-for-datavis/spotify.csv',index_col="Date",parse_dates=True)
spotify.head()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify)
```

```python
sns.set_style("dark") #darkgrid, whitegrid, dark, white, ticks
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify)
```
