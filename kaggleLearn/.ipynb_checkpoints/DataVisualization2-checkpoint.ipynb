# Univariate plotting with pandas

```python
import matplotlib.pyplot as plt
import pandas as pd
reviews = pd.read_csv('../../Data/wine-reviews/winemag-data_first150k.csv', index_col=0)
reviews.head(3)
```

```python
reviews['province'].value_counts().head(10).plot.bar()
```

```python
(reviews['province'].value_counts().head(10) / len(reviews)).plot.bar()
```

```python
reviews.points.value_counts().sort_index().plot.bar()
```

```python
reviews.points.value_counts().sort_index().plot.line()
```

```python
reviews.points.value_counts().sort_index().plot.area()
```

```python
reviews[reviews.price<200].price.plot.hist()
```

```python
reviews.price.plot.hist()
```

```python
reviews[reviews.price>1500]
```

```python
reviews.points.plot.hist()
```

```python
pd.set_option('max_columns', None)
pokemon = pd.read_csv('../../Data/pokemon/pokemon.csv')
pokemon.head(3)
```

```python
pokemon.type1.value_counts().plot.bar()
```

```python
pokemon.hp.value_counts().sort_index().plot.line()
```

```python
pokemon.weight_kg.hist()
```

# Bivariate plotting with pandas

```python
reviews.head()
```

## Scatter plot

```python
reviews[reviews.price < 100].sample(100).plot.scatter(x='price',y='points')
```

```python
reviews[reviews.price<100].plot.scatter(x='price',y='points')
```

## Hexplot

```python
reviews[reviews.price <100].plot.hexbin(x='price',y='points',gridsize=15)
```

## Stacked plots

```python
wine_counts = pd.read_csv('../../Data/most-common-wine-scores/top-five-wine-score-counts.csv',index_col=0)
wine_counts.head()
```

```python
wine_counts.plot.bar(stacked=True)
```

```python
wine_counts.plot.area()
```

## Bivariate line chart

```python
wine_counts.plot.line()
```
