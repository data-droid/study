```python
import pandas as pd
reviews = pd.read_csv('../../../Data/wine-reviews/winemag-data-130k-v2.csv',index_col=0)
reviews.head(3)
```

```python
from plotnine import *
```

```python
top_wines = reviews[reviews.variety.isin(reviews.variety.value_counts().head(5).index)]
top_wines.head(3)
```

## The grammar of graphics

```python
df = top_wines.head(1000).dropna()

(ggplot(df)
    + aes('points','price')
    + geom_point())
```

```python
df = top_wines.head(1000).dropna()

(ggplot(df)
    + aes('points','price')
    + geom_point()
    + stat_smooth()
)
```

```python
df = top_wines.head(1000).dropna()

(ggplot(df)
    + geom_point()
    + aes(color='points')
    + aes('points','price')
    + stat_smooth()
)
```

```python
df = top_wines.head(1000).dropna()

(ggplot(df)
     + aes('points', 'price')
     + aes(color='points')
     + geom_point()
     + stat_smooth()
     + facet_wrap('~variety')
)
```

```python
(ggplot(df)
    + geom_point(aes('points','price')))
```

```python
(ggplot(df, aes('points','price'))
    + geom_point())
```

## More plotnine

```python
(ggplot(top_wines)
    + aes('points')
    + geom_bar())
```

```python
(ggplot(top_wines)
    + aes('points','variety')
    + geom_bin2d(bins=20))
```

```python
(ggplot(top_wines)
    + aes('points','variety')
    + geom_bin2d(bins=20)
    + coord_fixed(ratio=1)
    + ggtitle("Top Five Most Common Wine Variety Points Awarded"))
```
